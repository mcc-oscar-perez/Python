```python
# podeos crear 2 listas 

# Creamos la lista x con los numeros del 1 al 100
x = list(range(1,101))

# Creamos la lista y con el cuadrado de cada numero en x
y = []

for i in x:
    y.append(i**2)

# graficamos x y y
plt.plot(x,y)
```

```python
# Creamos el grafico utilizando subplots
# fig = almacena la figura completa
# ax = almacena los ejes de la grafica

fig, ax = plt.subplots()
ax.plot(x,y)
```

```python
# Creamos un disccionario con tres platillos y su precio 
comidas = {"lasa√±a":350, "sopa":150, "roast beef":650}
```

```python
# Creamos un greafico de barras
fig,ax = plt.subplots()
ax.bar(comidas.keys(), comidas.values()) # colocamos primero el valor en eje x y luego el valor en eje y

# personalizamos el grafico
ax.set(
    title="Precios de comidas",
    xlabel ="Comidas",
    ylabel ="Precios"
)
```

```python
# Podemos crear un grafico pero con barras horizontales
fig,ax = plt.subplots()

ax.barh(list(comidas.keys()), list(comidas.values()))

ax.set(
    title = "Precios de comidas",
    xlabel = "Precios",
    ylabel = "Comidas"
)
```

```python
# Podemos crear una distribuci√≥n de 1000 valores aleatorios distribuidos normalmente
x = np.random.randn(1000) # Creamos aun array de mil numeros con puntos decimales entre 0 y 1  
# Creamos un un histograma 
fig, ax = plt.subplots()
ax.hist(x)
```

```python
# Podemos crear un afigura que contenga 4 sub-graficos
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2,figsize=(10,5))
```

```python
# Creamos graficos con conenido 

# Creamos el liensio en blanco primero 
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2,figsize=(10,5))

# Asignamos a cada valor de ax un grafico
ax1.plot(x1,y1)

ax2.scatter(x11,y11)

ax3.bar((comidas.keys()),comidas.values())

ax4.hist(np.random.randn(1000))
```
