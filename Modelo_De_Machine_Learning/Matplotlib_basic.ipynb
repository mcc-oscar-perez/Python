```python
# podeos crear 2 listas 

# Creamos la lista x con los numeros del 1 al 100
x = list(range(1,101))

# Creamos la lista y con el cuadrado de cada numero en x
y = []

for i in x:
    y.append(i**2)

# graficamos x y y
plt.plot(x,y)
```

```python
# Creamos el grafico utilizando subplots
# fig = almacena la figura completa
# ax = almacena los ejes de la grafica

fig, ax = plt.subplots()
ax.plot(x,y)
```

```python
# Pasos completos para el grafico de casos de COVID-19
x = list(range(1,101))
y = []

for i in x:
    y.append(i**2)

fig,ax =plt.subplots()

ax.plot(x,y)

# Podemos personalizar la grafica agregando etiquetas y un titulo
ax.set(
        title='Grafico de casos de COVID-19',
        xlabel="dias",
        ylabel="casos confirmados"
        )

# Podemos guardar la el grafico
fig.savefig("D:\\Cursos\\Python\\Modelo_De_Machine_Learning\\Pruebas\\grafico_covid19.png")
```

```python
# importamos numpy para crear datos de prueba
import numpy as np

# Creamos datos de prueba
x1 = np.linspace(0,100,20)
y1 = x1**2

# Creamos el grafico de dispersion
fig,ax = plt.subplots()
ax.scatter(x1,y1)

# personalizamos el grafico
ax.set(
    title="Grafico de dispersion",
    xlabel="eje x",
    ylabel="eje y"
)
```
