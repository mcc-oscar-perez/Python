# Modelo de Machine Learning que predice las probabilidades de sobrevivir a la trajedia del Titcanic

```python
# impotar librerias 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# importar librerias de Machine learning
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import metrics
from sklearn import tree 
```

```python
# cargar nuestra base de datos csv como un dataframe de pandas 
df = pd.read_csv("DataSet_Titanic.csv")
df
```

# Separamos la bse de datos
- x = valores predictores
- y = valores a predecir 

```python
# Separamos de la base de datos "X" los valores de la columna "Superviviente" al eleguir axis = 1 indicamos que "supervivientes es una columna"
X = df.drop('Sobreviviente',axis =1)

# Guardamos en la columna "Y" la columna "Superviviente" para, de esta manera x almacena todo el data frame menos esa columna que estara guarda en "Y"
Y = df.Sobreviviente
```

```python
Y.head()
```

# Creamos nuestro Calsificador de arbol de decisiones 
- **max_depth**, indicamos la profundida maxima de nuestros arbol, en este caso 2
- **random_state**, indicamos la aleatoriedad de nuestro arbol 

```python
# Creamos el objeto Arbol
Arbol = DecisionTreeClassifier(max_depth=2, random_state=42)
```

```python
# Entrenamos nuestro modelo 
Arbol.fit(X,Y)
```

# Entrenamos el modelo

```python
# Predicciones sobre nuestro dataframe
pred_Y = Arbol.predict(X) 

# Comparamos con las etiquetas reales 
print('Precision: ', accuracy_score(pred_Y,Y))
```

# Validacion estadistica

```python
# Matriz de confusión 
confusion = confusion_matrix(Y,pred_Y)
```

```python
# Matriz de confusión a partir de predicciones
disp = metrics.ConfusionMatrixDisplay.from_predictions(
                                                Y,       # Valores reales
                                                pred_Y,  # Valores predichos
                                                labels=[0, 1], # 
                                                display_labels=['No sobrevivió','Sobrevivió'], # Etiquetas
                                                normalize=None                       
                                                )
plt.show()
```
