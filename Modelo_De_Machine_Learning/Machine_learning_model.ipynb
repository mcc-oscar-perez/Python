# Modelo de Machine Learning que predice las probabilidades de sobrevivir a la tragedia del Titanic

```python
# importar librerías 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# importar librerías de Machine learning
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import metrics
from sklearn import tree 
```

```python
# cargar nuestra base de datos csv como un dataframe de pandas 
df = pd.read_csv("DataSet_Titanic.csv")
df
```

# Separamos la base de datos
- x = valores predictores
- y = valores a predecir 

```python
# Separamos de la base de datos "X" los valores de la columna "Superviviente" al elegir axis = 1 indicamos que "supervivientes es una columna"
X = df.drop('Sobreviviente',axis =1)

# Guardamos en la columna "Y" la columna "Superviviente" para, de esta manera x almacena todo el data frame menos esa columna que estará guarda en "Y"
Y = df.Sobreviviente
```

```python
X.head()
```

```python
Y.head()
```

# Creamos nuestro Clasificador de árbol de decisiones 
- **max_depth**, indicamos la profundidad máxima de nuestro árbol, en este caso 2
- **random_state**, indicamos la aleatoriedad de nuestro árbol 

```python
# Creamos el objeto Arbol
Arbol = DecisionTreeClassifier(max_depth=3, random_state=42)
```

```python
# Entrenamos nuestro modelo 
Arbol.fit(X,Y)
```

# Entrenamos el modelo

```python
# Predicciones sobre nuestro dataframe
pred_Y = Arbol.predict(X) 

# Comparamos con las etiquetas reales 
print('Precision: ', accuracy_score(pred_Y,Y))
```

# validación estadística

```python
# Matriz de confusión 
confusion = confusion_matrix(Y,pred_Y)
```

# Mostramos el modelo de árbol de decisión que utilizamos gráficamente

- En este árbol podemos ver como que se consideró para tomar la definición de si alguien viviría o moriría.
- mientras más profundidad colocamos en **"depth"** el árbol tomara o más o menos cosas en cuenta

```python
plt.Figure(figsize=(10,8))
tree.plot_tree(
                Arbol, # modelo 
                filled=True, # rellenamos el árbol de color
                feature_names=X.columns # asiganmos nombre a las columnas con las que hay en el dataframe X
                )
plt.show()
```

- **Edad**, aquí separa para saber si la persona es menor o igual a esa edad
- **gini**, aquí visualizamos que tan mezclado está el grupo, cuanto más bajo el valor más pura es y por lo tanto no hay mucha mezcla de clases vivo o muerto
- **Samples**, aquí vemos cuantas muestras o filas de la base de datos se tomaron en cuenta en esta rama
- **Value**, es el conteo por clase dentro del nodo lado izquierdo =  no sobrevivieron y derecho = sobrevivieron
- **class**, es la clase mayoritaria en el nodo 


```python
# Tamaño y nitidez de la figura
fig, ax = plt.subplots(figsize=(18, 12), dpi=150)

# Dibujo del árbol con fuente más grande y cajas redondeadas/coloreadas
tree.plot_tree(
    Arbol,
    ax=ax,
    feature_names=X.columns,                  
    class_names=['No sobrevivió','Sobrevivió'],
    filled=True,                              # rellena con color por clase
    rounded=True,                             # bordes redondeados
    fontsize=12,                              # tamaño de la fuente
    impurity=True,                            # muestra gini
    proportion=False                          # proporciones en value
)

fig.tight_layout()
plt.show()
```
